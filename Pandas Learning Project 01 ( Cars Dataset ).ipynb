# Data Analysis using Pandas Learning Project 01: Cars Dataset

* Finding Null values and filling them with mean of that column
* Checking different 'makers' and counting their occurance in the dataset
* Filtering records for 'Asia' and 'Europe'
* Removing all records where weight is over 4000
* Increasing all the values of a column by 3

```python
import pandas as pd
import numpy as np
```

```python
cars = pd.read_csv('Cars+Data.csv')
cars.head()
```

```python
cars.info()
```

# 1. Finding Null Values 7 Filling them with mean

```python
cars.isnull().sum()
```

```python
# Filling missing values with mean of the column

cars['Cylinders'].fillna(cars['Cylinders'].mean(), inplace=True)
cars['Horsepower'].fillna(cars['Horsepower'].mean(), inplace=True)
cars['MPG_City'].fillna(cars['MPG_City'].mean(), inplace=True)
cars['MPG_Highway'].fillna(cars['MPG_Highway'].mean(), inplace=True)
cars['Weight'].fillna(cars['Weight'].mean(), inplace=True)
cars['Wheelbase'].fillna(cars['Wheelbase'].mean(), inplace=True)
cars['Length'].fillna(cars['Length'].mean(), inplace=True)
cars['EngineSize'].fillna(cars['EngineSize'].mean(), inplace=True)
```

```python
cars.head()
```

# 2. Checking different 'makers' and counting

```python
cars['Make'].value_counts().head(10)
```

# 3. Filtering records for 'Asia' and 'Europe'

```python
cars[cars['Origin'].isin(['Asia', 'Europe'])]
```

# 4. Removing all records where weight is over 4000

```python
cars = cars[cars['Weight'] < 4000]
cars
```

# 5. Increasing all the values of a column by 3

```python
cars['MPG_City'] = cars['MPG_City'].apply(lambda x:x+3)
```
